# .. Descripción y Exploración de datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
```

```python
from warnings import simplefilter
simplefilter(action="ignore", category=RuntimeWarning)
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/MandbeZ/TFM_sequia/main/datos/spi_spei/indices_spi3.csv', sep = ',', parse_dates=True)
# df = df.loc[:,["fecha","101"]]
df.dropna(inplace = True)
df.set_index('fecha',inplace = True)
```

```python
df.plot(figsize=(15, 8))
```

# Prueba de estacionariedad de Dickey Fuller

```python
#Función para realizar la Prueba de Dickey-Fuller aumentada (  ADF augmented Dickey–Fuller)
def adf(x):
  res = adfuller(x)
  print("Test-Statistic:", res[0])
  print("P-Value:", res[1])
  if res[1] < 0.05:
    print("La serie es estacionaria")
  else:
    print("La serie es NO estacionaria")
```

```python
for i in df.columns.values:
    print(f'Estación:{i} ')
    adf(df[i])
```

# Prueba de Autocorrelación ACF  y Autocorrelación  Parcial PACF

```python
'''Autocorrelación ACF'''
for i in df.columns.values:
    fig, ax = plt.subplots(figsize=(12, 5))
    plt.suptitle('ACF - AUTOCORRELACIÓN ESTACIÓN '+str(i), fontsize=20, color = 'darkblue', weight='bold')
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plot_acf(df[i].dropna(), ax=ax,lags=40.,title='');
```

```python
'''Autocorrelación Parcial PACF'''
for i in df.columns.values:
    fig, ax = plt.subplots(figsize=(12, 5))
    plt.suptitle('PACF - AUTOCORRELACIÓN PARCIAL ESTACIÓN '+str(i), fontsize=20, color = 'darkblue', weight='bold')
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plot_pacf(df[i].dropna(), ax=ax,lags=40,title='');
```
